```python
#Importación de librerías necesarias
import pandas as pd
import seaborn as sns
from seaborn import lmplot
import io
import requests
import matplotlib.pyplot as plt

```

```python
#carga de datos de nuestro Dataset de notas de alumnos
url="https://github.com/estefaniateni/ProyectoFinalVisualizaci-n/blob/main/Maestro_Estadillo_Alumnos.xls?raw=true"
alumnos=pd.read_excel(url, sheet_name="Hoja1")
```

```python
print(alumnos)
alumnos.head(10)
alumnos.tail(10)
```

**Análisis exploratorio de datos mediantes técnicas de Visualización por Gráficos**

```python
#Contabilizamos el número de alumnos según el sexo
sns.countplot(x="SEXO", data=alumnos)
```

```python
#Contabilizamos el número de alumnos APROBADOS y SUSPENSOS
sns.countplot(x="APROBADO", data=alumnos)
```

```python
#GRAFICO DE DISPERSIÓN 
#cruzamos datos de MEDIA CON FALTAS ACUMULADAS

lmplot('MEDIA', 'FALTAS ACUMULADAS', data=alumnos, fit_reg=False)
```

```python
lmplot(x='MEDIA', y='FALTAS ACUMULADAS', hue='SEXO', data=alumnos)
```

```python
#Contabilizamos el número de alumnos por edad
sns.countplot(x="EDAD", data=alumnos)
```

```python
#Realizamos un histograma por edades
sns.displot(data=alumnos, x="EDAD", kde=True)
```

```python
#Realizamos un histograma por EDADES y clasificando por SEXO
sns.displot(data=alumnos, x="EDAD", hue="SEXO", kind="kde")
```

```python
#Realizamos un HISTOGRAMA DE BARRAS APILADAS por EDADES y clasificando por SEXO
sns.displot(data=alumnos, x="EDAD", hue="SEXO", multiple="stack")
```

```python

#Contabilizamos el número de alumnos por ETAPA (INFORMÁTICA/ADMINISTRACIÓN)
sns.countplot(x="CURSO", data=alumnos)
```

```python
#Contabilizamos número de alumnos por AÑO
sns.countplot(y="PERIODO", data=alumnos, orient="h")
```

```python
#Contabilizamos el número de alumnos por año dividiendo entre Hombres y Mujeres
sns.countplot(y="PERIODO", hue="SEXO", data=alumnos, orient="h")
```

```python
#BOXPLOT
#BoxPlot de la edad por sexo y turno
with sns.axes_style(style='ticks'):
  ax = sns.catplot("SEXO", "EDAD", "TURNO" ,data=alumnos, kind="box", palette="Set3")
```

```python
sns.catplot(x="CURSO", y="MEDIA", hue="SEXO", data=alumnos)
```

```python
#Gráfico de dispersion
#Distribución cruzada de Edad y Nota (solo datos de mujeres)

sns.jointplot(x='MEDIA',y='EDAD',data=alumnos[(alumnos.SEXO == "M")])
```

```python
#Gráfico de dispersion
#Distribución cruzada de Edad y Nota, agrupado por SEXO
sns.jointplot(data=alumnos, x="MEDIA", y="EDAD", hue="SEXO")
```

```python
sns.jointplot(data=alumnos, x="MEDIA", y="FALTAS ACUMULADAS", hue="SEXO", kind="hist")
```

```python
sns.jointplot(data=alumnos, x="MEDIA", y="FALTAS ACUMULADAS")
```

```python
sns.jointplot(data=alumnos, x="EDAD", y="FALTAS ACUMULADAS", hue="SEXO")
```

```python
#Distribución de la edad según categorías:survived y clase

#Cambiamos el font
sns.set(font_scale=1)

#FacetGrid - Construir una matriz de gráficos
g = sns.FacetGrid(alumnos, row='SEXO',col='ETAPA')
g.map(sns.distplot, "EDAD")

plt.show()
```

```python
#Distribución de la edad según categorías:survived y clase

#Cambiamos el font
sns.set(font_scale=1)

#FacetGrid - Construir una matriz de gráficos
g = sns.FacetGrid(alumnos, row='SEXO',col='APROBADO')
g.map(sns.distplot, "EDAD")

plt.show()
```

```python
#Distribución de la edad según categorías:SEXO y APROBADO

#Cambiamos el font
sns.set(font_scale=1)

#FacetGrid - Construir una matriz de gráficos
g = sns.FacetGrid(alumnos, row='SEXO',col='CURSO', hue='APROBADO')
g.map_dataframe(sns.histplot,x='SEXO')

plt.show()
```
